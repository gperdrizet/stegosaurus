# Text encoding

## 1. Notebook set-up

```python
message = "Hello, Stegosaurus!"
```

## 2. String to binary encoding

```python
def encode_message(message: str) -> list[int]:
    '''Encode a string message as a flat list of bits (UTF-8, MSB first).'''

    # Collector for encoded result
    bits = []

    for byte in message.encode('utf-8'):
        for i in range(7, -1, -1):
            bits.append((byte >> i) & 1)

    return bits
```

```python
encoded = encode_message(message)
print(f'Encoded message: {encoded}')
```

## 3. Binary to string decoding

```python
def decode_message(bits: list[int]) -> str:
    '''Decode a flat list of bits (UTF-8, MSB first) back to a string.'''

    if len(bits) % 8 != 0:
        raise ValueError(f'Bit list length must be a multiple of 8, got {len(bits)}')

    byte_array = bytearray()

    for i in range(0, len(bits), 8):

        byte = 0

        for bit in bits[i:i + 8]:
            byte = (byte << 1) | bit
    
        byte_array.append(byte)

    return byte_array.decode('utf-8')
```

```python
decoded = decode_message(encoded)
print(f'Decoded message: {decoded}')
```
